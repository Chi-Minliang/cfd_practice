### Read csv results simulated by C++ implementation

```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("../build/sod.csv", "r") as file:
    data = [v for v in csv.reader(file)]
data = np.array(data[1:], dtype=np.double)

plt.figure(figsize=(10,8), dpi=100)
states = (r"$\rho$", r"$u$", r"$p$", r"$\rho e$")
x = data[:, 0]
profiles = (data[:, 1], data[:, 2], data[:, 3], data[:, 4])
for i, (state, profile) in enumerate(zip(states, profiles)):
    plt.subplot(2, 2, i+1)
    plt.plot(x, profile, linewidth=1)
    plt.xlabel("$x$")
    plt.ylabel(state)
plt.tight_layout()
```
